```python
import numpy as np
import pandas as pd
import scipy

import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from IPython.display import Image 
from pydot import graph_from_dot_data
from sklearn import metrics


#a bunch of imports not all are necessary
```

```python
url = "https://raw.githubusercontent.com/daztakiff/DiscoverAI_Projects_Group-15-respository/main/lab2_Takiff/UnsupervisedLearning/Custom/healthy_lifestyle_city_2021.csv"

#na_values changes all '-' characters to NaN 
df = pd.read_csv(url, na_values='-')
df.head()
```

```python
df.columns
#print columns so you can access them when u need to later
```

```python
#replace all values (£, %) in specified columns with nothing (clean data)
df['Cost of a bottle of water(City)'] = df['Cost of a bottle of water(City)'].str.replace('£','')
df['Obesity levels(Country)'] = df['Obesity levels(Country)'].str.replace('%','')
df['Cost of a monthly gym membership(City)'] = df['Cost of a monthly gym membership(City)'].str.replace('£','')
```

```python
#drop catagorical data that doesn't affect result
df = df.drop('City', 1)
```

```python
#drops all NaN data from your dataset
df = df.dropna()
#cleans up indexes since you just removed rows
df = df.reset_index()
```

```python
#clumps 10 groups of three effectivally splitting rank in three tiers 0 (best rank), 1(medium rank), 2(lower rank)
for i in range(3):
  for j in range(10):
    df['Rank'][i*10 + j] = i

#wasnt evenly divisible by three put last entry in group 2
df['Rank'][30] = 2
df
```

```python
#Set up X variables we use to predict Y
X = df.drop(columns = ['Rank'])
# Set up Y as variable we want to predict
Y = df['Rank']
```

```python
#train model (can be done in different ways)
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=1)
```

```python
#load tree and fit to our model
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
#make predictions
y_pred = dt.predict(X_test)
```

```python
#some code I found for accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#some code I found for displaying tree
import graphviz 
from sklearn import tree
dot_data = tree.export_graphviz(dt, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("tree")
```
