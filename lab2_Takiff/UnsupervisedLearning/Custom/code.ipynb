# Clean data

Earn Your Wings

```python
# Like in lab 1, use the MinMax scaler to scale values for better accuracy
# drop everything but life expectnacy and pollution
rows_to_drop = ["City", "Rank", "Sunshine hours(City)","Cost of a bottle of water(City)", "Obesity levels(Country)", "Annual avg. hours worked", "Happiness levels(Country)", "Outdoor activities(City)", "Number of take out places(City)", "Cost of a monthly gym membership(City)"]
for i in rows_to_drop:
    df = df.drop(i, axis=1)

from sklearn.preprocessing import MinMaxScaler
cols = df.columns
ms = MinMaxScaler()

df = ms.fit_transform(df)
df = pd.DataFrame(df, columns=[cols])

#just something I wanted to see
df
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt # for data visualization
# Define a range for possible k values. In this example, we choose 1 to 10
k_rng = range(1,10)

# Declare an array to store the values from the sum of squared error values.
sse = []

# Using a for loop, go through each value in the k range and compute the sse value
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Life expectancy(years) (Country)','Pollution(Index score) (City)']])
    sse.append(km.inertia_)

# Plot the sse value for each k.
# Notice that 3 is the elbow value.
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

Implement a K-Means Clustering algorithm on the cleaned data set. Use the elbow method to find the right value of k to use. Add comments in your code to explain each step that you take in your implementation.

```python
# Define a KMeans object with 5 as the value of k. Call the object km
km = KMeans(n_clusters=5)

# Predict the data set excluding the name column. 
# Save the prediction in an array and call it y_predicted
df.rename(columns={'Life expectancy(years) (Country)': 'lifeExpect', 'Pollution(Index score) (City)': 'Pollution'})
print(df)

y_predicted = km.fit_predict(df[['lifeExpect','Pollution']])
# Add a column to the data frame to store the predicted cluster ID of each element
df['cluster'] = y_predicted

df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]


# Plot the data frames with different colors to differentiate them
plt.scatter(df1.lifeExpect,df1['Pollution'],color='green')
plt.scatter(df2.lifeExpect,df2['Pollution'],color='red')
plt.scatter(df3.lifeExpect,df3['Pollution'],color='black')
plt.scatter(df4.lifeExpect,df4['Pollution'],color='purple')
plt.scatter(df5.lifeExpect,df5['Pollution'],color='yellow')

# Define the labels on the x and y axes
plt.xlabel('lifeExpect')
plt.ylabel('Pollution')

# Plot the legend
plt.legend()
```
