Earn Your Wings

Use a decision tree classifier on the cleaned data set to predict y for the given data. Report the accuracy score for 30 runs. Add comments in your code to explain each step that you take in your implementation.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# defining our features and labels
X = train_df.drop(columns = ['Rank', 'index', 'City', 'Obesity levels(Country)', 'Cost of a bottle of water(City)', 'Life expectancy(years) (Country)', 'Pollution(Index score) (City)', 'Annual avg. hours worked', 'Happiness levels(Country)', 'Outdoor activities(City)', 'Cost of a monthly gym membership(City)'])
y = train_df['Happiness levels(Country)']
# changing the datatype from Object to numeric values
X=X.astype('float')
y=y.astype('int')

# allocation 20% of data to test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
# initializing our model
model = DecisionTreeClassifier()
#training our model
model.fit(X_train, y_train)
predictions = model.predict(X_test)

score = accuracy_score(predictions, y_test)
score
```

The accuracy score is 57% after 30 runs.
